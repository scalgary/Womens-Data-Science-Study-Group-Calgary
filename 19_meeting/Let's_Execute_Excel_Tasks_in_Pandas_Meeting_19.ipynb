## Data clean and EDA

```
df.rename({' Income ':'Income'}, axis=1, inplace=True) # Rename Income column removing white space
df['Income'] = df['Income'].str.replace('$','').str.replace(',','').astype(float) # remove '$', replace comma and convert the data type to float
```

```
df.info()
```

```
df.head()
```

```
# Column summary
df['Income'].describe().to_frame()
```

```
# numeric columns summary
df.describe()
```

### Applying formulas on the columns

```
df['Total_purchases'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df['Children'] = df['Kidhome'] + df['Teenhome']
```

```
df.head()
```

```
# Create a categorical variable using list comprehensions 
df['Children_cat'] = ['Yes' if x > 0 else 'No' for x in df['Kidhome']+df['Teenhome']]
df.head()
```

### Sum of columns presented in columns

```
# First, create a sum for the 'Kidhome' and 'Teenhome'
sum_row = df[["Kidhome","Teenhome"]].sum()
sum_row
```

To perform this action, it is necessary to transpose the data and convert the Series to a DataFrame. We use the T function to switch the data from row to column.

```
df_sum = pd.DataFrame(data=sum_row).T
df_sum
```

### Sorting sheet based in one or more columns

```
sorted_by_Income = df.sort_values(['Income'], ascending=False)
sorted_by_Income
```

### Display top 10 Income

```
sorted_by_Income["Income"].head(10)
```

```
sorted_by_Income['Income'].head(10).plot(kind="barh")
plt.show()
```

### Get the mean of a particular column

```
df["Income"].mean()
```

### Select a subset of the dataset

```
df_subset = df[['Education', 'Income']]
df_subset.head()
```

### Pivot table with grouping on Education and summarization on the sum of Income.


```
income_by_education = df_subset.pivot_table(index=['Education'])
income_by_education
```

### Data visualization from pivot table

```
income_by_education.plot()
plt.show()
```

### Pivot with multiple columns

```
df_subset = df[['Education', 'Marital_Status', 'Income']]
df_subset.head()
```

```
income_by_edu_marital = df_subset.pivot_table(index=['Education', 'Marital_Status'])
income_by_edu_marital
```

```

income_by_edu_marital.plot(kind='bar', figsize=(20,8))
plt.show()
```

### Make table long to wide

```
df_subset = df[['Education', 'Marital_Status','Country','Income']]
df_subset.head()
```

```
# Sum of Income by Education level Marital status by Country
df_subset.pivot_table(index=['Education','Marital_Status'], columns='Country', values='Income',aggfunc='sum')
```

```
# Add the sum margins to the previews table
df_subset.pivot_table(index=['Education','Marital_Status'], columns='Country', values='Income',
                    margins=True,  # add margins
                    aggfunc='sum')  # sum margins (rows/columns)
```

```
df_subset = df[['Education', 'Marital_Status','Income']]

df_subset.groupby(['Education', 'Marital_Status']).mean().reset_index()
```

```
df_subset.pivot_table(index=['Education', 'Marital_Status'], 
                    aggfunc=['min','max', 'mean','first'])
```

```
# groupby equivalent
df_subset.groupby(['Education', 'Marital_Status']).agg(['min','max', 'mean','first'])
```
